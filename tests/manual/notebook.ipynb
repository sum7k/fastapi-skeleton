```python
from models.domain import UserDTO
from core.database import get_db_session_context, init_db
from services.auth import TokenService
import uuid
```

```python
# Initialize database and create session properly

await init_db()
print("Database initialized successfully!")
```

```python
try:
    async with get_db_session_context() as session:
        print("Session created successfully")
        token_service = TokenService(session)
        print("TokenService created")

        user_dto = UserDTO(id=uuid.uuid4())
        print(f"UserDTO created with ID: {user_dto.id}")

        # Test the token creation
        token = await token_service.create_access_token(user_dto)
        print(f"Generated token: {token}")
        print(f"Token length: {len(token)}")

except Exception as e:
    print(f"Error occurred: {type(e).__name__}: {str(e)}")
    import traceback

    traceback.print_exc()
```
